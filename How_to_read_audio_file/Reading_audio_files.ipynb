**Librosa is a python library, which is used for signal processing**

**When we load an audio_file by default it returns sample rate and floating values of audio signal**

```python
import matplotlib.pyplot as plt
import numpy as np
import librosa
import librosa.display
import IPython.display as ipd # To listen music or Audio
```

```python
# load audio file

music_ = r"C:\Users\hp\3D Objects\Audio_files\uploading_audio_files\sample-6s.wav"
```

```python
# Listening audio_file

ipd.Audio(music_)
```

```python
music, sr1 = librosa.load(music_)  # sr----> sample range , by default it will take 22050(sr) and mono paramter will be dedfault
```

```python
music 
```

```python
sr1
```

```python
music.size # no.of.samples 
```

```python
music.shape
```

```python
# duration of one sample

sample_dur = 1 / sr1

print(f"duration of one sample : {sample_dur: .6f} seconds")   # .6f is 6 decimal points
```

```python
# duration of the audio signal in seconds

dur = sample_dur * len(music)
print(f"duration of signal is : {dur: .2f} seconds")
```

```python
# visualize the waveforms
# waveforms are going to be normalized between (-1 to +1)

plt.figure(figsize = (7,3))
librosa.display.waveshow(music, alpha = 0.5)
plt.title("Test_Music")
plt.ylim(-1, 1)    # (range of y_axis)
```
